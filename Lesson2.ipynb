```python
# Создадим массив с заданными параметрами
t = np.linspace(0, 10, 51)
f = np.cos(t)
# Нарисуем графие f(t) зеленого цвета
plt.plot(t, f, color="green")
# Доавим название для графика
plt.title("График f(t)")
# Добавим название осей и ограничение по осям
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

**Задание 4** 

```python
plt.style.use('fivethirtyeight')
# Считываем данные с файла в дата фрейм
df = pd.read_csv("Desktop\RootFolder\Data_Science\creditcard.csv")
# считаем кол-во для поля Class
df_class = df["Class"].value_counts()
# строим столбчатую диаграму
df_class.plot(kind="barh")
plt.show()
```

```python
# логарифмируем её
df_class.plot(kind="barh")
plt.xscale(value="log")
plt.show()
```

```python
# обычные операции
first = df[df["Class"] == 0].loc[:, "V1"]
# мошеннические
second = df[df["Class"] == 1].loc[:, "V1"]
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.hist(first, bins=20, alpha=0.5, density=True, color="grey", label="Class 0")
ax.hist(second, bins=20, alpha=0.5, density=True, color="red", label="Class 1")
plt.ylabel("Class")
plt.legend()
plt.show()
```
